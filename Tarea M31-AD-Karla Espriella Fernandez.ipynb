```python
#agregar columna en donde en base al pais se agregara la zona horaria 
import pytz
import pycountry
from countryinfo import CountryInfo # type: ignore
from country_capitals import get_capital

def get_country_code(country_name):
        try:
            return pycountry.countries.lookup(country_name).alpha_2
        except LookupError:
            return None
def Obtener_capital(country_name):
    try:
         return get_capital(country_name)
    except:
         return None
          
def ZonaHoraria(country_name):
     try:
          return pytz.country_timezones[country_name]
     except:
          return None
     
#obtengo la capital de cada pais
df['Capital']=df['Nationality'].apply(lambda x:Obtener_capital(x))

#obtengo el codigo de cada pais 
df['country_code'] = df['Nationality'].apply(lambda x:get_country_code(x))

#con el codigo de cada pais obtengo la zona horaria
df['Zonahoraria']=df['country_code'].apply(lambda x:ZonaHoraria(x))

print(df['country_code'])
print(df['Capital'])
print(df['Zonahoraria'])

```
